```python
#Fibonacci Sequence: Generate the first \(n\) terms.
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
```

```python
#Two Sum Problem: Find two numbers in a list that sum to a specific target.
def two_sum(nums, target):
    lookup = {}
    for i, num in enumerate(nums):
        if target - num in lookup:
            return [lookup[target - num], i]
        lookup[num] = i
```
